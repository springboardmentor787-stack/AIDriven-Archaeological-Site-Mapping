```python
import os
import cv2
import numpy as np

input_folder = "/content/dataset/train/images"
output_folder = "/content/processed_dataset/images"

os.makedirs(output_folder, exist_ok=True)

img_size = (224,224)

for img in os.listdir(input_folder):

    img_path = os.path.join(input_folder,img)

    image = cv2.imread(img_path)

    if image is None:
        continue

    resized = cv2.resize(image,img_size)

    normalized = resized/255.0

    normalized_img = (normalized*255).astype(np.uint8)

    cv2.imwrite(os.path.join(output_folder,img),normalized_img)

print("Resizing and normalization completed")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

image_path="/content/processed_dataset/images"
label_path="/content/processed_dataset/labels"

images=os.listdir(image_path)

train_imgs,temp_imgs=train_test_split(images,test_size=0.3,random_state=42)

val_imgs,test_imgs=train_test_split(temp_imgs,test_size=0.5,random_state=42)
```

```python
def move_files(files,img_dest,lbl_dest):

    for img in files:

        label=img.replace(".jpg",".txt")

        shutil.copy(os.path.join(image_path,img),
                    os.path.join(img_dest,img))

        shutil.copy(os.path.join(label_path,label),
                    os.path.join(lbl_dest,label))


move_files(train_imgs,
"/content/final_dataset/train/images",
"/content/final_dataset/train/labels")

move_files(val_imgs,
"/content/final_dataset/val/images",
"/content/final_dataset/val/labels")

move_files(test_imgs,
"/content/final_dataset/test/images",
"/content/final_dataset/test/labels")

print("Dataset split completed")
```

```python
import cv2
import os
import numpy as np

input_path = "/content/dataset/train/images"
output_path = "/content/processed/images"

os.makedirs(output_path, exist_ok=True)

for img in os.listdir(input_path):

    img_path = os.path.join(input_path, img)

    image = cv2.imread(img_path)

    if image is None:
        continue

    resized = cv2.resize(image, (224,224))

    normalized = resized / 255.0

    final_img = (normalized * 255).astype(np.uint8)

    cv2.imwrite(os.path.join(output_path, img), final_img)

print("Resize and normalization completed")
```

```python
from sklearn.model_selection import train_test_split

image_path = "/content/processed/images"
images = os.listdir(image_path)

train_imgs, temp_imgs = train_test_split(images, test_size=0.3, random_state=42)
val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

print("Train:", len(train_imgs))
print("Validation:", len(val_imgs))
print("Test:", len(test_imgs))
```
