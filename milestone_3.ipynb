```python
import os
import cv2
import numpy as np
import pandas as pd
from IPython.display import Image, display
from google.colab import drive

# Mount Google Drive if not already mounted
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# Folder path: **Please update this path to your actual image *folder* in Google Drive**
image_folder = "/content/drive/MyDrive/runs/detect/exp2/img12_jpg.rf.3ad1a53865aeec70fc9a6b92a9a23401.jpg"

# ◰ Get first image from folder
image_list = [img for img in os.listdir(image_folder) if img.endswith(('.jpg','.png','.jpeg'))]

if image_list:
    img_name = image_list[0]   # select first image
    path = os.path.join(image_folder, img_name)

    print("Selected Image:", img_name)

    # Display image
    display(Image(filename=path))

    # Read image
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Feature extraction
    mean_intensity = np.mean(gray)
    variance = np.var(gray)

    edges = cv2.Canny(gray,100,200)
    edge_density = np.sum(edges) / edges.size

    # Create dataframe
    df = pd.DataFrame([[mean_intensity, variance, edge_density]],
                      columns=["brightness","texture","edge_density"])

    print(df)

else:
    print("No images found in folder!")
```
